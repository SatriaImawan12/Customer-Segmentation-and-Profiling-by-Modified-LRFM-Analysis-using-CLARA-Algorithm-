# **3. Outlier**

```python
import pandas as pd

# Kelompokkan data berdasarkan cluster_label
grouped = df_profiling.groupby('cluster_label')

# Definisikan DataFrame untuk menyimpan hasil outlier counts
outlier_df = pd.DataFrame(columns=['Cluster', 'Frequency Outliers', 'Monetary Outliers', 'Discounts Outliers', 'User Age Outliers', 'Total Outliers'])

# Lakukan perhitungan outlier
for name, group in grouped:
    outliers = {
        'frequency': (group['frequency'].quantile(0.75) + 1.5 * (group['frequency'].quantile(0.75) - group['frequency'].quantile(0.25))),
        'monetary': (group['monetary'].quantile(0.75) + 1.5 * (group['monetary'].quantile(0.75) - group['monetary'].quantile(0.25))),
        'discounts': (group['discounts'].quantile(0.75) + 1.5 * (group['discounts'].quantile(0.75) - group['discounts'].quantile(0.25))),
        'user_age': (group['user_age'].quantile(0.75) + 1.5 * (group['user_age'].quantile(0.75) - group['user_age'].quantile(0.25)))
    }

    # Agregasi discounts per pengguna unik
    discounts_per_user = group.groupby('user_id')['discounts'].sum()

    outlier_counts = {
        'frequency': group[group['frequency'] > outliers['frequency']].shape[0],
        'monetary': group[group['monetary'] > outliers['monetary']].shape[0],
        'discounts': discounts_per_user[discounts_per_user > outliers['discounts']].shape[0],
        'user_age': group[group['user_age'] > outliers['user_age']].shape[0]
    }

    # Hitung total outlier untuk setiap cluster
    total_outliers = sum(outlier_counts.values())

    # Menambahkan hasil perhitungan ke DataFrame
    outlier_df.loc[len(outlier_df)] = {
    'Cluster': name,
    'Frequency Outliers': outlier_counts['frequency'],
    'Monetary Outliers': outlier_counts['monetary'],
    'Discounts Outliers': outlier_counts['discounts'],
    'User Age Outliers': outlier_counts['user_age'],
    'Total Outliers': total_outliers
}


# Tampilkan DataFrame
display(outlier_df)
```

Karena terdapat outlier pada data profiling, ukuran pemusatan yang paling tepat untuk merepresentasikan karakteristik data profiling adalah median. Berikut alasannya.

* Median adalah nilai tengah dari data yang telah diurutkan.
* Median tidak terpengaruh oleh nilai-nilai ekstrem atau outlier dalam data.

Oleh karena itu, median memberikan representasi yang lebih akurat dari "pusat" data ketika ada outlier.

**notes:** Mean sangat terpengaruh oleh outlier, karena outlier dapat menggeser mean ke arah nilai ekstrem.
